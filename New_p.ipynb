```python
def pivott(df, FP, FV, PE):
    data=df
    for i in range(0,len(FP)):
        filt = FP[i]
        Var = FV[i]
        for j in Var:
            data = data[data[filt] == j]
    Result = dict()
    for i in data[PE]:
        Result[i] = Result.get(i, 0) + 1
    sorted_keys = sorted(Result, key=Result.get, reverse=True)
    sorted_Result=dict()
    for w in sorted_keys:
        sorted_Result[w] = Result[w]
    return sorted_Result
```

```python
path = "filter.txt"
with open(path) as f:
    filt = f.read()
filt=filt.replace('\n',"").replace(" ","").replace("]}","").replace("%"," ")
filt=filt.replace(']','').split(";")
if len(filt[-1])==0:
    filt.pop()
    
Main_result={}
for line in filt:
    filt_p=[]
    filt_v=[]
    result=[]
    line = line.split(',')
    if line[0].split('{')[0]=='Filter':
        lin=line[0].replace('Filter{[','')
        lin=lin.split('&')
        
        for k in lin:
            f_V=[]
            m=k.split('=')
            f_V.append(m[1])
            filt_p.append(m[0])
            filt_v.append(f_V)
        PE=line[1].split('=')[1]
        Heading=line[2].split('=')[1].replace('}','')
        Main_result[Heading]=pivott(df1, filt_p, filt_v, PE)
        
    elif line[0].split('{')[0]=='Generate':
        report_name=line[0].split('{')[1].replace('Filename=','').replace('}','')+'.pdf'
        print('Generating ',report_name,'..........')
        pdf = FPDF()
        for Heading,data in Main_result.items():
            pdf.add_page()
            pdf.set_xy(0, 0)
            pdf.set_font('arial', 'B', 12)
            pdf.cell(60)
            pdf.cell(75, 10, Heading, 0, 2, 'C')
            pdf.cell(90, 10, " ", 0, 2, 'C')
            pdf.cell(-40)
            pdf.cell(40, 10, 'S.No', 1, 0, 'C')
            pdf.cell(50, 10, 'Param', 1, 0, 'C')
            pdf.cell(40, 10, 'Counts', 1, 2, 'C')
            pdf.cell(-90)
            pdf.set_font('arial', '', 10)
            keyss= list(data.keys())
            values = list(data.values())
            for i in range(0, len(keyss)):
                if i==10:
                    break
                pdf.cell(40, 8, '%s' % (str(i)), 1, 0, 'C')
                pdf.cell(50, 8, '%s' % (str(keyss[i])), 1, 0, 'C')
                pdf.cell(40, 8, '%s' % (str(values[i])), 1, 2, 'C')
                pdf.cell(-90)
            
            
            xlabel('Type')
            ylabel('Counts')
            i=6
            if len(keyss)<6:
                i=len(keyss)
            plt.bar(list(data.keys())[0:i], list(data.values())[0:i], width=0.5, color="#91eb87")
            savefig('chart.png')
            plt.close()
            
            pdf.cell(90, 10, " ", 0, 2, 'C')
            pdf.image('chart.png', x = None, y = None, w = 0, h = 0, type = '', link = '')
        pdf.output(report_name, 'F')
            

    else:
        print('Syntax is wrong')
```
